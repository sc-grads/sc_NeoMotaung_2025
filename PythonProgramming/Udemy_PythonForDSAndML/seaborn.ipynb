```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips') #seaborn has some pre-installed datasets
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'])
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
plt.show()
```

```python
sns.rugplot(tips['total_bill'])
plt.show()
```

```python
#kde = kernel destination estimation plot
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
plt.show()
```

```python
sns.countplot(x='sex', data=tips)
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex', split=True)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips) #swarmplots dont scale well with large data
#These two combined are better for analysis purposes rather tham presentations
plt.show()
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='violin') #General form
plt.show()
```
